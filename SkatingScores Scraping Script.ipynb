```python
base = "https://skatingscores.com"
event_urls = ["https://skatingscores.com/1819/wc/", "https://skatingscores.com/2021/wc/", "https://skatingscores.com/2122/wc/",
             "https://skatingscores.com/1819/gpusa/", "https://skatingscores.com/1819/gpcan/", "https://skatingscores.com/1819/gpfin/",
              "https://skatingscores.com/1819/gpjpn/", "https://skatingscores.com/1819/gprus/", "https://skatingscores.com/1819/gpfra/",
              "https://skatingscores.com/1819/gpf/", "https://skatingscores.com/1819/ec/", "https://skatingscores.com/1819/4cc/",
              "https://skatingscores.com/1819/lomtro/", "https://skatingscores.com/1819/nebtro/", "https://skatingscores.com/1819/gldspn/",
              "https://skatingscores.com/1920/lomtro/", "https://skatingscores.com/1920/nebtro/", "https://skatingscores.com/1920/warcup/",
              "https://skatingscores.com/1920/gldspn/", "https://skatingscores.com/1920/gpusa/", "https://skatingscores.com/1920/gpcan/",
             "https://skatingscores.com/1920/gpfra/", "https://skatingscores.com/1920/gpchn/", "https://skatingscores.com/1920/gprus/",
             "https://skatingscores.com/1920/gpjpn/", "https://skatingscores.com/1920/gpf/", "https://skatingscores.com/1920/ec/",
             "https://skatingscores.com/1920/4cc/", "https://skatingscores.com/2122/lomtro/", "https://skatingscores.com/2122/nebtro/",
             "https://skatingscores.com/2122/warcup/", "https://skatingscores.com/2122/gldspn/", "https://skatingscores.com/2122/gpusa/",
             "https://skatingscores.com/2122/gpcan/", "https://skatingscores.com/2122/gpita/",
             "https://skatingscores.com/2122/gpfra/", "https://skatingscores.com/2122/gprus/", "https://skatingscores.com/2122/ec/",
             "https://skatingscores.com/2122/4cc/", "https://skatingscores.com/2122/oly/", "https://skatingscores.com/2122/olyt/"]
urls = []
s = requests.Session()
retries = Retry(total=5,
            backoff_factor=0.1,
            status_forcelist=[ 500, 502, 503, 504 ])
s.mount('https://', HTTPAdapter(max_retries=retries))
for url in event_urls:
    response = s.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    tabs = soup.find_all(class_ = "col-12 col-t-6 col-d-6")
    for i in range(4):
        table = tabs[i].find_all("tr")
        try:
            panel = base+table[1].find_all("a")[1]["href"]
            sp = base+table[2].find_all("a")[0]["href"]
            fp = base+table[2].find_all("a")[1]["href"]
            new_list = [panel, sp, fp]
            urls.append(new_list)
        except IndexError:
            continue
print(urls)
```
